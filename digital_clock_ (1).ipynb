```python
from playsound import playsound
from tkinter import *
from tkinter import messagebox
import time
import datetime
import pygame 

from tkinter import *
import pygame
import time

pygame.mixer.init()

frame = Tk()
frame.geometry('750x550')
frame.title('Countdown Timer')

def clock():
    current_time = time.strftime('%H:%M:%S %p')
    l6.config(text=current_time)
    l6.after(1000, clock)

def timer():
    try:
        hour = int(t1.get())
        minute = int(t2.get())
        second = int(t3.get())
    except ValueError:
        messagebox.showerror("Invalid input", "Please enter valid numbers.")
        return

    total_seconds = hour * 3600 + minute * 60 + second

    def countdown(total_seconds):
        if total_seconds > 0:
            mins, secs = divmod(total_seconds, 60)
            hours, mins = divmod(mins, 60)
            timer_display = f"{hours:02}:{mins:02}:{secs:02}"
            l9.config(text=timer_display)
            frame.after(1000, countdown, total_seconds - 1)
        else:
            l9.config(text="Time's up!")
            pygame.mixer.music.load('rooster2.mp3')
            pygame.mixer.music.play(loops=0)

    countdown(total_seconds)

def stop():
    pygame.mixer.music.stop()

current_time = datetime.date.today()
heading = Label(frame, text='Clock', font=('Verdana', 39), fg='black', bg='white')
l1 = Label(frame, text='Enter the hour: ')
l7 = Label(frame, text='Enter the minutes: ')
l8 = Label(frame, text='Enter the second: ')
l2 = Label(frame, text=current_time, font=(18))
l3 = Label(frame, text=f" Month: {current_time.strftime('%B')}", font=('Arial', 18))
l4 = Label(frame, text=f" Day: {current_time.strftime('%A')}", font=('Arial', 18))
l5 = Label(frame, text=f" Year: {current_time.year}", font=('Arial', 18))
l6 = Label(frame, font=('Verdana', 39))
l9 = Label(frame, font=('Verdana', 24), fg='red')  # Timer display
b1 = Button(frame, text='Start', command=timer)
b2 = Button(frame, text='Stop', command=stop)
t1 = Entry(frame, width=30)
t2 = Entry(frame, width=30)
t3 = Entry(frame, width=30)

heading.pack()
l6.pack()
l3.pack()
l4.pack()
l5.pack()
l1.pack()
t1.pack()
l7.pack()
t2.pack()
l8.pack()
t3.pack()
l2.pack()
b1.pack()
b2.pack()
l9.pack()  # Pack the timer display
clock()
frame.mainloop()
```
